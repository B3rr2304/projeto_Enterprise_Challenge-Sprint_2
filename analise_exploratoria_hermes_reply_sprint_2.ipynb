```python
#Instalação de pacotes
import pandas as pd
import matplotlib.pyplot as plt

#Lendo os dados obtidos
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("todos_os_dados_sensores.csv")

#Mostrar as 5 primeiras linhas
print("Visualizando os primeiros dados:")
print(df.head())

#Estatísticas básicas
print("\nEstatísticas básicas:")
print(df.describe())

#Gráfico: Temperatura, Umidade e Potenciômetro ao longo do tempo
plt.figure(figsize=(12,6))

plt.plot(df["Amostra"], df["Temperatura (°C)"], label="Temperatura (°C)", marker="o")
plt.plot(df["Amostra"], df["Umidade (%)"], label="Umidade (%)", marker="x")
plt.plot(df["Amostra"], df["Potenciometro"], label="Potenciômetro", marker="s")

plt.title("Dados coletados dos sensores")
plt.xlabel("Amostra")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
